```python
import gradio as gr
```

```python
def greet(text):
    print(f"shout has been called with input{text}")
    return text.upper()

greet ("example text")
```

```python
simple_chat = gr.Interface(fn=greet, inputs="text", outputs="text", flagging_mode = "never")
```

```python
simple_chat.launch(share = True)
```
